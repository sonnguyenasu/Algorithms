```python
class Person(object):
    def __init__(self,name):
        self.name = name
        self.friends = []
        
    #declare the connection
    def addfriend(self,other):
        if other in self.friends:
            print(other.name + " has been a friend of" + self.name +" already")
            
        else:
            self.friends.append(other)
            other.friends.append(self)
            print(other.name + " and " + self.name + " has become friends.")
            
    #remove the connection        
    def unfriend(self,other):
        if other not in self.friends:
            print("Oh no",other.name,"is not",self.name,"'s friend.")
        else:
            self.friends.remove(other)
            other.friends.remove(self)
    
    #print the object
    def __str__(self):
        res = self.name + "'s friend(s): "
        for it in range(len(self.friends)-1):
            res+=self.friends[it].name + ", "
        if self.friends:
            res += self.friends[-1].name
        return res
    
    #===============BREADTH FIRST SEARCH================#
    def BFS(self):
        i = 1
        level = {self:0}
        parent = {self:None}
        frontiers = [self]
        while frontiers:
            nextr = []
            for u in frontiers:
                for v in u.friends:
                    if v not in level:
                        level[v] = i
                        parent[v] = u
                        nextr.append(v)
            frontiers = nextr
            i += 1
        return (level,parent)
    
    #===============USING BFS Figuring out shortest path=============#
    def shortestPath(self,other):
        res = self.BFS()
        parent = res[1]
        level = res[0][other]
        temp = other
        for it in range(level):
            print(temp.name, end = '-> ')
            temp = parent[temp]
        print(self.name)
            
```

```python
#class community contain the whole group of people with such declared connection
class Community(object):
    def __init__(self):
        self.group = {}
    def newMember(self):
        name = input("His/her name: ")
        while name in self.group:
            name = input("His/her name: ")
        temp = Person(name)
        number_of_friend = int(input("Number of friends: "))
        while number_of_friend > len(self.group):
            number_of_friend = int(input("Number of friends: "))
        for it in range(number_of_friend):
            friend_name = input(">> ")
            if friend_name in self.group.keys():
                temp.addfriend(self.group[friend_name])
            else:
                while friend_name not in self.group.keys():
                    print("The person is not in this community yet.")
                    friend_name = input(">> ")
                temp.addfriend(self.group[friend_name])
        self.group[name] = temp
    def __str__(self):
        res = "Our community has " + str(len(self.group)) + " people:\n"
        for key,value in self.group.items():
            res += value.__str__()
            res += "\n"
        return res

        
                
```

```python
com = Community()
for it in range(8):
    com.newMember()
print(com)
```
